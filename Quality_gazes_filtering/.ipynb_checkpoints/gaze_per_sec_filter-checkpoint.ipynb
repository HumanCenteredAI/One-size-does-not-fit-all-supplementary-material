# DATA FILTER BASED ON QUALITY OF GAZE

## Extract features needed for filtering out low quality gazes

```python
df = pd.read_csv('demo_m.csv')
```

```python
df
```

```python
df1 = df[['Participant', 'GazesMean', 'TextRTMean']]
```

```python
df1
```

```python
#number of instances before filtering
num_instances = len(df1)
print ("number of instances before filtering = ", num_instances)
```

## Create 2 new files containing instances to be removed and to be kept

```python
#participants who need to be removed based on having low quality gaze
rows_to_remove = df1.query('TextRTMean < 1 or GazesMean/TextRTMean < 5')

#number of instances to filter
num_instances_remove = len(rows_to_remove)
print ("number of instances to filter out = ", num_instances_remove)
```

```python
# #create a new file containing instances needed to be removed
# rows_to_remove.to_csv("instances_to_remove.csv", index=False)
```

```python
#participants who wouldn't be removed
rows_to_keep = df1.query('TextRTMean >= 1 and GazesMean/TextRTMean >= 5')

#number of instances to keep
num_instances_keep = len(rows_to_keep)
print ("number of paragraphs to keep = ", num_instances_keep)
```

```python
# #create a new file containing instances to be kept
# rows_to_keep.to_csv("instances_to_keep.csv", index=False)
```

## Read nlp_gaze_fixation file and keep instances containing high quality gazes 

```python
#read files
keep_df = pd.read_csv("instances_to_keep.csv")
nlp_gaze_fixation_df = pd.read_csv("NLP__gazes_fixation_z.csv")
```

```python
#number of participants for nlp_gaze_fixation_df
num_participants_nlp = nlp_gaze_fixation_df.Participant.nunique()
print ("number of unique participants in nlp_gaze_fixation_df = ", num_participants_nlp)

#number of instances to keep
num_instances_nlp = len(nlp_gaze_fixation_df)
print ("number of instances in nlp_gaze_fixation_df = ", num_instances_nlp)
```

```python
filter_df = pd.merge(nlp_gaze_fixation_df, keep_df[['Participant']], on=['Participant'], how = 'inner')
filter_df
```

```python
#number of participants after filtering data
num_participants_after_filter = filter_df.Participant.nunique()
print ("number of unique participants in filtered data = ", num_participants_after_filter)

#number of instances after filtering data
num_instances_after_filter = len(filter_df)
print ("number of instances in filtered data = ", num_instances_after_filter)
```
